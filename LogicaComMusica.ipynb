```python
#Definimdo número máximo 
numeroMaximo = 10

contagens = [
    "é um, é um, é",
    "é dois, é dois, é",
    "é três, é três, é",
    "é quatro, é quatro, é",
    "é cinco, é cinco, é cinco, é",
    "é seis, é seis, é seis, é",
    "é sete, é sete, é",
    "é oito, é oito, é",
    "é nove, é nove, é",
    "é dez, é dez, é dez, é dez, é"
]

#Criando um loop
i = 1
while i <= numeroMaximo:

    print("Mariana conta", i)
    

    print("Mariana conta", i, ",", contagens[i-1])
    

    print("Ana, viva a Mariana, viva a Mariana\n")
    
    #atribuindo mais um 
    i += 1
```

```python
patinhos = 5

while patinhos > 0:
    print(f"{patinhos} patinhos foram passear")
    print("Além das montanhas")
    print("Para brincar")
    print("A mamãe gritou: Quá, quá, quá, quá")
    patinhos -= 1
    if patinhos == 0:
        print("Nenhum patinho voltou de lá")
    else:
        print(f"Mas só {patinhos} patinhos voltaram de lá")
    print()

print("A mamãe patinha foi procurar")
print("Além das montanhas")
print("Na beira do mar")
print("A mamãe gritou: Quá, quá, quá, quá")
print("E os cinco patinhos voltaram de lá")
```

```python

elefantes = 1
while elefantes <= 10:
    if elefantes == 1:
        print(f"{elefantes} elefante incomoda muita gente")
    else:
        print(f"{elefantes} elefantes incomodam, {'incomodam, ' * (elefantes - 1)}incomodam muito mais")
    elefantes += 1


elefantes = 10
while elefantes >= 1:
    if elefantes == 1:
        print(f"{elefantes} elefante incomoda muito menos")
    else:
        print(f"{elefantes} elefantes incomodam, {'incomodam, ' * (elefantes - 1)}muito menos")
    elefantes -= 1
```
